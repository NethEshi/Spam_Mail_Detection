```python
import pandas as pd #for data manipulation and analysis
import numpy as np #for calculation
from google.colab import files
uploaded=files.upload()
```

```python
df=pd.read_csv("spam.csv")
df.head()
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns',inplace=True)  #remone unwanted columns
```

```python
df.head()
```

```python
df.groupby('v1').describe() #grouping them based on v1
```

```python
df['Spam']=df['v1'].apply(lambda x:1 if x== 'spam' else 0) #use lamda function to adjust 0 and 1
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer #Ml can't understand Text. convert text in to numbers
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.v2,df.Spam,test_size=0.25) #split data set as 75-training , 25-testing
```

```python
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values) # need to vectorize the strings to a binary sequence and use that for the model training type(X_train_count)
```

```python
type(X_train_count.toarray())
```

```python
print(X_train_count[10]) #how many similar words repeated in the msg
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count.toarray(),y_train)
```

```python
email = ['Happy Birthday','Congratulations. You have won 10000 USD USD USD.' , 'moda Eshi.']


email_count=v.transform(email)
model.predict(email_count)
```
